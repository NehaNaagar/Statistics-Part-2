# Theoretical:-

1. What is hypothesis testing in statistics?

- Hypothesis testing is a statistical technique used to make inferences about a population based on a sample of data. It involves formulating a hypothesis, collecting and analyzing data, and determining whether the data supports or rejects the hypothesis.



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

- The null hypothesis (H0) is a statement of no effect or no difference, while the alternative hypothesis (H1) is a statement of an effect or difference. The null hypothesis is tested against the alternative hypothesis, and if the data supports the alternative hypothesis, the null hypothesis is rejected.



3. What is the significance level in hypothesis testing, and why is it important?

- The significance level (α) is the probability of rejecting the null hypothesis when it is true. It's a critical value that determines whether the results are statistically significant. A common significance level is 0.05, which means there's a 5% chance of rejecting the null hypothesis when it's true.



4. What does a P-value represent in hypothesis testing?

- The P-value represents the probability of observing the results (or more extreme) assuming the null hypothesis is true. A small P-value indicates that the results are unlikely to occur by chance, and the null hypothesis can be rejected.



5. How do you interpret the P-value in hypothesis testing?

- If the P-value is less than the significance level (α), the results are statistically significant, and the null hypothesis can be rejected. If the P-value is greater than α, the results are not statistically significant, and the null hypothesis cannot be rejected.


6. What are Type 1 and Type 2 errors in hypothesis testing?

- Type 1 error occurs when the null hypothesis is rejected when it's true (false positive). Type 2 error occurs when the null hypothesis is not rejected when it's false (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- one-tailed test is used when the alternative hypothesis is directional (e.g., H1: μ > 0), while a two-tailed test is used when the alternative hypothesis is non-directional (e.g., H1: μ ≠ 0).

8. What is the Z-test, and when is it used in hypothesis testing?

- The Z-test is a statistical test used to compare the mean of a sample to a known population mean. It's used when the sample size is large, and the population standard deviation is known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

- The Z-score is calculated as (x̄ - μ) / (σ / √n), where x̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size. The Z-score represents the number of standard deviations from the population mean.


10. What is the T-distribution, and when should it be used instead of the normal distribution?

- The T-distribution is a probability distribution used when the sample size is small, and the population standard deviation is unknown. It's used instead of the normal distribution when the sample size is less than 30.


11. What is the difference between a Z-test and a T-test?

- The Z-test is used when the population standard deviation is known, while the T-test is used when the population standard deviation is unknown.


12. What is the T-test, and how is it used in hypothesis testing?

- The T-test is a statistical test used to compare the mean of a sample to a known population mean or to compare the means of two samples. It's used when the sample size is small, and the population standard deviation is unknown.


13. What is the relationship between Z-test and T-test in hypothesis testing?

- The Z-test and T-test are both used for hypothesis testing, but the Z-test is used when the population standard deviation is known, while the T-test is used when the population standard deviation is unknown.

14. What is a confidence interval, and how is it used to interpret statistical results?

- A confidence interval is a range of values within which a population parameter is likely to lie. It's used to interpret statistical results by providing a margin of error.

15. What is the margin of error, and how does it affect the confidence interval?

- The margin of error is the maximum amount by which the sample estimate may differ from the true population parameter. It affects the confidence interval by determining its width.


 16. How is Bayes' Theorem used in statistics, and what is its significance?

- Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It's significant because it allows us to incorporate prior knowledge and new data to make more accurate predictions and decisions. Bayes' Theorem is widely used in machine learning, data analysis, and scientific research.


17. What is the Chi-square distribution, and when is it used?

- The Chi-square distribution is a probability distribution used to model the sum of squared standard normal variables. It's used in hypothesis testing, particularly in tests for categorical data, such as goodness of fit tests and tests for independence.

18. What is the Chi-square goodness of fit test, and how is it applied?

- The Chi-square goodness of fit test is used to determine whether a set of observed frequencies differs significantly from a set of expected frequencies. It's applied by calculating the Chi-square statistic and comparing it to a critical value or using a P-value to determine significance.


19. What is the F-distribution, and when is it used in hypothesis testing?

- The F-distribution is a probability distribution used to model the ratio of two independent Chi-square variables. It's used in hypothesis testing, particularly in tests for comparing variances and in analysis of variance (ANOVA).


20. What is an ANOVA test, and what are its assumptions?

- An ANOVA test is used to compare the means of three or more groups. Its assumptions include normality of the data, homogeneity of variance, and independence of observations.


21. What are the different types of ANOVA tests?

- The different types of ANOVA tests include:

-One-way ANOVA

-Two-way ANOVA

-Repeated measures ANOVA

-Factorial ANOVA


22. What is the F-test, and how does it relate to hypothesis testing?

- The F-test is used to compare the variances of two or more groups. It's related to hypothesis testing because it's used to determine whether the observed differences between groups are statistically significant. The F-test is often used in ANOVA.

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy import stats

# Set the seed for reproducibility
np.random.seed(0)

# Define the parameters of the population
population_mean = 5
population_std = 2

# Define the sample size
n = 100

# Generate the sample data
sample_data = np.random.normal(population_mean, population_std, n)

# Calculate the sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)

# Define the null and alternative hypotheses
null_hypothesis = "The population mean is equal to 5"
alternative_hypothesis = "The population mean is not equal to 5"

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results
print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is sufficient evidence to suggest that the population mean is not equal to 5")
else:
    print("Fail to reject the null hypothesis")
    print("There is insufficient evidence to suggest that the population mean is not equal to 5")

```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats

# Generate a dataset
np.random.seed(0)
data = np.random.normal(5, 2, 100)

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

# Calculate the standard error
standard_error = sample_std / np.sqrt(len(data))

# Calculate the critical value for the 95% confidence interval
critical_value = stats.t.ppf(0.975, len(data) - 1)

# Calculate the margin of error
margin_of_error = critical_value * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the confidence interval
print(f"95% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

# Interpret the result
print("Interpretation:")
print("We are 95% confident that the true population mean lies within the interval.")
print(f"The interval is ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f}).")
print("This means that if we were to repeat the sampling process many times,")
print("we would expect the true population mean to lie within this interval 95% of the time.")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level):
    # Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(len(sample_data))

    # Calculate the critical value
    critical_value = stats.t.ppf((1 + confidence_level) / 2, len(sample_data) - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval

# Example usage
sample_data = np.random.normal(5, 2, 100)
confidence_level = 0.95
confidence_interval = calculate_confidence_interval(sample_data, confidence_level)
print(f"Confidence Interval: {confidence_interval}")


```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    prior (float): Prior probability of the hypothesis.
    likelihood (float): Likelihood of the data given the hypothesis.
    evidence (float): Evidence for the hypothesis.

    Returns:
    float: Posterior probability of the hypothesis.
    """
    # Calculate the posterior probability
    posterior = (prior * likelihood) / evidence

    return posterior

# Example usage
prior = 0.5  # Prior probability
likelihood = 0.8  # Likelihood of the data given the hypothesis
evidence = 0.6  # Evidence for the hypothesis

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior:.4f}")


```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np

def calculate_expected_frequencies(contingency_table):
    # Calculate the row and column totals
    row_totals = np.sum(contingency_table, axis=1)
    column_totals = np.sum(contingency_table, axis=0)

    # Calculate the grand total
    grand_total = np.sum(contingency_table)

    # Calculate the expected frequencies
    expected_frequencies = np.outer(row_totals, column_totals) / grand_total

    return expected_frequencies

# Define the contingency table
contingency_table = np.array([[10, 20], [30, 40]])

# Calculate the expected frequencies
expected_frequencies = calculate_expected_frequencies(contingency_table)

# Print the expected frequencies
print(f"Expected Frequencies:\n{expected_frequencies}")


```

```python

# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA+

import numpy as np
from scipy import stats

def check_anova_assumptions(data):
    # Check normality using the Shapiro-Wilk test
    normality_results = []
    for i in range(len(data)):
        _, p_value = stats.shapiro(data[i])
        normality_results.append(p_value > 0.05)

    # Check independence using the Durbin-Watson test
    independence_results = []
    for i in range(len(data)):
        residuals = data[i] - np.mean(data[i])
        dw_statistic = 2 * (1 - np.sum(residuals[:-1] * residuals[1:]) / np.sum(residuals ** 2))
        independence_results.append(dw_statistic > 1.5 and dw_statistic < 2.5)

    # Check equal variance using Levene's test
    equal_variance_results = stats.levene(*data)

    return normality_results, independence_results, equal_variance_results

# Example usage
data = [np.random.normal(0, 1, 100), np.random.normal(1, 1, 100), np.random.normal(2, 1, 100)]
normality_results, independence_results, equal_variance_results = check_anova_assumptions(data)
print("Normality Results:", normality_results)
print("Independence Results:", independence_results)
print("Equal Variance Results:", equal_variance_results)

```

```python

# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
from scipy import stats

# Define the data
np.random.seed(0)
data1 = np.random.binomial(1, 0.5, 100)
data2 = np.random.binomial(1, 0.6, 100)

# Perform a Z-test for comparing proportions
p1 = np.mean(data1)
p2 = np.mean(data2)
se = np.sqrt(p1 * (1 - p1) / len(data1) + p2 * (1 - p2) / len(data2))
z_statistic = (p1 - p2) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
    print("The proportions are significantly different")
else:
    print("Fail to reject the null hypothesis")
    print("The proportions are not significantly different")
```

```python

# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# Define the expected frequencies under the null hypothesis
expected_frequencies = np.array([100, 200, 300, 200, 100, 100])

# Perform a Chi-square test for goodness of fit
observed_frequencies, _ = np.histogram(data, bins=6, range=(-3, 3))

# Scale the expected frequencies to match the observed frequencies sum
expected_frequencies = expected_frequencies * np.sum(observed_frequencies) / np.sum(expected_frequencies)

chi2_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# Print the results
print("Chi-square statistic:", chi2_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
    print("The data does not fit the expected distribution")
else:
    print("Fail to reject the null hypothesis")
    print("The data fits the expected distribution")

# Visualize the results
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=6, range=(-3, 3), alpha=0.5, label='Observed')
plt.bar(range(6), expected_frequencies, alpha=0.5, label='Expected')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of the Data')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(6), observed_frequencies, label='Observed')
plt.plot(range(6), expected_frequencies, label='Expected')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Line Plots of the Data')
plt.legend()

plt.tight_layout()
plt.show()
```
